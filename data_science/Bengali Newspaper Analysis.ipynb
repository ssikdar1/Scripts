```python
"""
Part of an experiment of mine to do some data analysis on the text
of articles on a Bengali Newspaper, to help figure out what are the
most important words to learn in order to become literate.

The newspaper of choice is anandabazar
link: http://www.anandabazar.com/

Step 1: Create the corpus
Step 2: Get the frequency of each word using Spark
"""

import requests
from bs4 import BeautifulSoup
import re
```

```python
# Functions to help create the corpus

def get_categories(url):
    """ Get the child paths for different categories from the navbar
    """
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)

    # Use beautiful soup to get all the navbar links
    filtered = soup.find(attrs = {'class': 'collapse navbar-collapse mob-nav'})
    subjects = {}
    for tag in filtered:
        if isinstance(tag, bs4.element.Tag):
            for anchor in tag.find_all('a'):
                href = anchor['href']
                if 'javascript' not in href: # dont want javascript void(0)
                    if href not in subjects:
                        subjects[href] = 1

    return subjects.keys()

    
def get_category_stories(category_child_path):
    """ Get the urls for the different stories for a given category
    """
    story_list = []
    
    # if a subject partial link /calcutta?ref=
    # the full link is http://www.anandabazar.com/calcutta?ref=
    page_url = "{}{}".format(url, child_path)

    resp = requests.get(page_url)
    page_soup = BeautifulSoup(resp.text)

    # grab all the links to the stories listed on the page in divs with class=sectionstoryheading
    stories = page_soup.find_all(attrs = {'class': 'sectionstoryheading'})
    for story in stories:
        story_list.append(story.find_all('a')[0]['href'])

    return story_list

def parse_page(page_url):
    """ parse the page return title and page text """
    resp = requests.get(page_url)
    page_soup = BeautifulSoup(resp.text)
    
    page_text = page_soup.find(attrs = {"class": "articleBody"}).find(attrs = {"id": "textbody"}).text

    # Get the title
    title = None
    title_search = page_soup.findAll(text=re.compile('storyTitleBe'))
    m = re.search(".+\'storyTitleBe\'\s*\:\s*\'(.+)\'.+", title_search[0])
    if m:
        title = m.groups()[0]
    
    return {
            'title': title,
            'page_text': page_text
    }
```

```python

def create_corpus():
    # The main webpage of the newspaper
    url = 'http://www.anandabazar.com'
    
    count = 0
    
    # get the files into one text file
    with open("/home/shan/Downloads/anandabazar.txt", "w+") as f:
        categories = get_categories(url)
        for category_path in categories:
            stories = get_category_stories(category_path)
            for story_url in stories:
                page_url = "{}{}".format(url, story_url)
                ret = parse_page(page_url)
                f.write(ret['title'].encode("UTF-8"))
                f.write(ret['page_text'].encode("UTF-8"))
                count += 1
        print count
create_corpus()
```

```python
"""
Now use the spark word count example to get the frequency of each word in the corpus
"""

from pyspark.sql import SparkSession
from operator import add
spark = SparkSession\
        .builder\
        .appName("AnnadaBazarWordCount")\
        .getOrCreate()
text_file = "/home/shan/Downloads/anandabazar.txt"

# split by space, each word count of 1, reduce by key
# swap k,v reduce by key, swap k,v back
lines = spark.read.text(text_file).rdd.map(lambda r: r[0])

counts = lines.flatMap(lambda x : x.split(' ')) \
        .map(lambda x: (x,1)) \
        .reduceByKey(add) \
        .map(lambda (x,y) : (y,x)) \
        .sortByKey(False) \
        .map(lambda (x,y) : (y,x))
        
output = counts.collect()
```

```python
# write to a file
with open('/home/shan/Downloads/anandabazar_frequency.txt', 'wb+') as f:
    for (word, count) in output:
        f.write("%s: %i\n" % (word.encode("UTF-8"), count))
```
